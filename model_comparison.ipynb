# Model analysis

```python
# sentiment 
# sentiment 
from textblob import TextBlob
def textblob_sen(x):
    score = TextBlob(x).sentiment[0]
    if score > 1:
        return 1
    elif score == 0:
        return 0
    else:
        return -1
        
        
from textblob import TextBlob
models = {'TextBlob': lambda x: textblob_sen(x)}

for name, model in models.items():
    df_test[name] = df_test['Tweet'].apply(lambda x: model(x))

df_test

```

```python
from sentistrength import PySentiStr

def sentistrength_sen(x):
    score = PySentiStr().getSentiment(x)
    if score > 1:
        return 1
    elif score == 0:
        return 0
    else:
        return -1
        
        
models = {'sentistrength': lambda x: sentistrength_sen(x)}

for name, model in models.items():
    df_test[name] = df_test['Tweet'].apply(lambda x: model(x))

df_test
```
